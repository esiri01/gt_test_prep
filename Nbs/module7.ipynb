### Question 16
Recall that the Box-Muller method takes two i.i.d. $\text{Unif}(0,1)$ PRNs, $U_1$ and $U_2$, and magically transforms them into two i.i.d. $\text{Nor}(0,1)$'s, $Z_1$ and $Z_2$. Specifically,  
$$ Z_1 = \sqrt{-2 \ln(U_1)} \cos(2\pi U_2) \quad \text{and} \quad Z_2 = \sqrt{-2 \ln(U_1)} \sin(2\pi U_2). $$  
It also turns out (I think I mentioned this in class) that $T = Z_2/Z_1$ has the Cauchy distribution. Using the B-M expressions for $Z_1$ and $Z_2$ above, find a nice simple expression for the Cauchy random variable $T$.

### Question 20
