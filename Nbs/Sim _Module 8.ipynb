## Q1
We are interested in seeing if the number of emergency department visits each day at the Georgia Tech clinic is Binomial(5, 0.5). Results over 200 days:

| # visits | # days |
|:--------:|:------:|
| 0 | 6 |
| 1 | 30 |
| 2 | 61 |
| 3 | 62 |
| 4 | 34 |
| 5 | 7 |

We will perform a 95% \( \chi^2 \) goodness-of-fit test.

### (a) How many intervals will you use?

**Solution**:

Expected counts under \( H_0 \):

$$
E_i = 200 \binom{5}{i} (0.5)^5
$$

The expected counts are:

| \( i \) | \( O_i \) | \( E_i \) |
|:-------:|:---------:|:---------:|
| 0 | 6 | 6.25 |
| 1 | 30 | 31.25 |
| 2 | 61 | 62.5 |
| 3 | 62 | 62.5 |
| 4 | 34 | 31.25 |
| 5 | 7 | 6.25 |

Since all expected counts \( \geq 5 \), we use **6 intervals**.

---

### (b) What is your test statistic value?

**Solution**:

Compute:

$$
\chi_0^2 = \sum_{i=0}^5 \frac{(O_i - E_i)^2}{E_i}
$$

Numerically:

$$
\chi_0^2 = 0.432
$$

---

### (c) What is your conclusion?

**Solution**:

At 5% significance, critical value:

$$
\chi_{0.05,5}^2 = 11.07
$$

Since:

$$
0.432 < 11.07
$$

we **fail to reject** \( H_0 \). Conclusion: **assume Binomial(5,0.5)**.

## Q2
Consider the PRNs: \( 0.18, 0.92, 0.61, 0.33 \).  
Use the Kolmogorov-Smirnov test for \( U(0,1) \).

### (a) What is the value of the test statistic?

**Solution**:

Order them: \( 0.18, 0.33, 0.61, 0.92 \)

| \( i \) | \( R_{(i)} \) | \( \frac{i}{n} - R_{(i)} \) | \( R_{(i)} - \frac{i-1}{n} \) |
|:-------:|:------------:|:--------------------------:|:---------------------------:|
| 1 | 0.18 | 0.07 | 0.18 |
| 2 | 0.33 | 0.17 | 0.08 |
| 3 | 0.61 | 0.14 | 0.11 |
| 4 | 0.92 | 0.07 | 0.18 |

Thus:

- \( D^+ = 0.17 \)
- \( D^- = 0.18 \)
- \( D_n = \max(D^+, D^-) = 0.18 \)

---

### (b) Do the PRNs pass the K-S test at \( \alpha = 0.05 \)?

**Solution**:

Critical value:

$$
D_{0.05,4} = 0.624
$$

Since:

$$
0.18 < 0.624
$$

we **fail to reject** uniformity.

## Q3
If \( X, Y \) are i.i.d. \( \text{Exp}(1) \), find \( \mathbb{E}[X^2 Y^3] \).

**Solution**:

Because \( X \) and \( Y \) are independent:

$$
\mathbb{E}[X^2 Y^3] = \mathbb{E}[X^2] \mathbb{E}[Y^3]
$$

Now:

$$
\mathbb{E}[X^2] = 2 \quad (\text{Gamma}(k=3,\lambda=1))
$$
$$
\mathbb{E}[Y^3] = 6 \quad (\text{Gamma}(k=4,\lambda=1))
$$

Thus:

$$
\mathbb{E}[X^2 Y^3] = 2 \times 6 = 12
$$

## Q4
Suppose \( X_1, X_2, X_3 \) are i.i.d. from \( \text{Normal}(\mu, \sigma^2) \), and:

\( X_1 = 1, X_2 = 5, X_3 = 3 \).

Find the MLE of \( \sigma^2 \).

**Solution**:

Compute:

- Mean:

$$
\bar{X} = 3
$$

- Sum of squares:

$$
(1-3)^2 + (5-3)^2 + (3-3)^2 = 4 + 4 + 0 = 8
$$

MLE of \( \sigma^2 \) (not unbiased):

$$
\widehat{\sigma}^2 = \frac{8}{3} \approx 2.67
$$

## Q5
TRUE or FALSE: If \( X_i \sim \text{Bernoulli}(p) \), then sample variance \( S^2 \) is unbiased for \( p^2 \).

**Solution**:

FALSE.

Sample variance is unbiased for \( p(1-p) \), **not** \( p^2 \).

---

## Q6
If \( X_1, \ldots, X_n \sim \text{Exp}(1/9) \), what is \( \mathbb{E}[S^2] \)?

**Solution**:

Variance of Exp(1/9) is:

$$
\sigma^2 = 9^2 = 81
$$

Thus:

$$
\mathbb{E}[S^2] = \sigma^2 = 81
$$

## Q7
TRUE or FALSE:  
The square root of the sample variance is an unbiased estimator for \( \sigma \).

**Solution**:

FALSE.

Even though \( \mathbb{E}[S^2] = \sigma^2 \), in general:

$$
\mathbb{E}[S] \neq \sigma
$$

## Q8
Suppose \( U_1 = 0.6 \), \( U_2 = 0.9 \), \( U_3 = 0.1 \).  
Use K-S with \( \alpha = 0.05 \).

**Solution**:

Order:

| \( i \) | 1 | 2 | 3 |
|:-------:|:-:|:-:|:-:|
| \( R_{(i)} \) | 0.1 | 0.6 | 0.9 |

| \( i/n - R_{(i)} \) | 0.233 | 0.067 | 0.1 |
| \( R_{(i)} - (i-1)/n \) | 0.1 | 0.267 | 0.233 |

Thus:

- \( D^+ = 0.233 \)
- \( D^- = 0.267 \)
- \( D = \max(D^+, D^-) = 0.267 \)

Critical value:

$$
D_{0.05,3} = 0.708
$$

Since:

$$
0.267 < 0.708
$$

we **fail to reject** uniformity.

## Q9
The number of defects follows Geometric(\( p \)):

| Defects \( x \) | Frequency |
|:---------------:|:---------:|
| 1 | 24 |
| 2 | 18 |
| 3 | 12 |
| 4 | 9 |
| 5 | 7 |

Find the MLE for \( p \).

**Solution**:

First:

$$
\bar{X} = \frac{(1)(24) + (2)(18) + (3)(12) + (4)(9) + (5)(7)}{70} = 2.3857
$$

Thus:

$$
\hat{p} = \frac{1}{\bar{X}} \approx 0.4192
$$

## Q1
Suppose you observe the following number of arrivals at a clinic over 120 days:

| # Arrivals | Frequency |
|:----------:|:---------:|
| 0 | 10 |
| 1 | 25 |
| 2 | 40 |
| 3 | 30 |
| 4 | 12 |
| 5 | 3 |

You suspect arrivals follow a Poisson distribution with unknown \( \lambda \).

(a) Estimate \( \lambda \) using MLE.

(b) Perform a 95% chi-square goodness-of-fit test. How many degrees of freedom should you use?

---

### Solution:

(a) MLE of \( \lambda \) for Poisson is the sample mean:

\[
\bar{X} = \frac{(0)(10) + (1)(25) + (2)(40) + (3)(30) + (4)(12) + (5)(3)}{120}
= \frac{0 + 25 + 80 + 90 + 48 + 15}{120} = \frac{258}{120} \approx 2.15
\]

Thus, \( \hat{\lambda} = 2.15 \).

---

(b) Degrees of freedom:

- Original number of cells = 6.
- Estimate \( \lambda \) ⇒ lose 1 d.f.
- So degrees of freedom = \( 6 - 1 - 1 = 4 \).

---

## Q2
Suppose \( X_1, X_2, X_3 \sim \text{Uniform}(0, \theta) \) and you observe: \( X_1 = 4 \), \( X_2 = 7 \), \( X_3 = 2 \).

Find the MLE for \( \theta \).

---

### Solution:

MLE for \( \theta \) under \( \text{Uniform}(0, \theta) \) is:

\[
\hat{\theta} = \max(X_1, X_2, X_3) = 7
\]

Thus, \( \hat{\theta} = 7 \).

---

## Q3
Suppose \( X \sim \text{Exponential}(\lambda) \).

Using method of moments, derive an estimator for \( \lambda \).

---

### Solution:

For exponential:

\[
\mathbb{E}[X] = \frac{1}{\lambda}
\]

Thus, setting sample mean \( \bar{X} \) equal to \( \mathbb{E}[X] \):

\[
\hat{\lambda}_{\text{MoM}} = \frac{1}{\bar{X}}
\]

Same as the MLE.

---

## Q4
You conduct a chi-square goodness-of-fit test on a geometric distribution.  
You have 5 intervals after combining small probability cells. You estimate \( p \) from the data.

How many degrees of freedom does your test statistic have?

---

### Solution:

Degrees of freedom:

\[
k - 1 - s = 5 - 1 - 1 = 3
\]

Thus, 3 degrees of freedom.

---

## Q5
Suppose \( X_1 = 8 \), \( X_2 = 10 \), \( X_3 = 9 \) are i.i.d. Normal(\( \mu, \sigma^2 \)).

Find the MLE for \( \mathbb{E}[X_i^2] \).

---

### Solution:

MLE by invariance:

\[
\mathbb{E}[X_i^2] = \mu^2 + \sigma^2
\]

First:

\[
\bar{X} = \frac{8+10+9}{3} = 9
\]

Then:

\[
S^2 = \frac{(8-9)^2 + (10-9)^2 + (9-9)^2}{3} = \frac{1+1+0}{3} = \frac{2}{3}
\]

Thus:

\[
\widehat{\mathbb{E}}[X_i^2] = 9^2 + \frac{2}{3} = 81 + \frac{2}{3} = 81.\overline{6}
\]

---

## Q6
TRUE or FALSE:  
The sample variance \( S^2 \) is an unbiased estimator for the variance of \( X \).

---

### Solution:

**TRUE.**

By construction:

\[
\mathbb{E}[S^2] = \sigma^2

## Q7
Suppose \( U_1 = 0.15 \), \( U_2 = 0.4 \), \( U_3 = 0.85 \). Use the Kolmogorov-Smirnov test at \( \alpha = 0.05 \).

Compute the K-S statistic.

---

### Solution:

Order:

\( 0.15, 0.4, 0.85 \)

Now:

| \( i \) | \( R_{(i)} \) | \( \frac{i}{n} - R_{(i)} \) | \( R_{(i)} - \frac{i-1}{n} \) |
|:-------:|:------------:|:--------------------------:|:---------------------------:|
| 1 | 0.15 | 0.183 | 0.15 |
| 2 | 0.4 | 0.267 | 0.067 |
| 3 | 0.85 | 0.15 | 0.516 |

Thus:

- \( D^+ = 0.267 \)
- \( D^- = 0.516 \)

Thus:

\[
D_n = 0.516
\]

Compare to critical value:

\[
D_{0.05,3} = 0.708
\]

Since \( 0.516 < 0.708 \), **fail to reject**.

---

## Q8
Suppose you have data believed to follow Weibull(\( \lambda, k \)), where both parameters are unknown.  
How many degrees of freedom do you lose in a chi-square goodness-of-fit test?

---

### Solution:

- Estimate 2 parameters ⇒ lose 2 degrees of freedom.
- So d.f. = number of intervals \( - 1 - 2 \).
