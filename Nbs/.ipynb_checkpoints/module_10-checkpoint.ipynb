## Question 1: Antithetic Random Numbers  
(Lesson 10.7: Antithetic Random Numbers)  
Suppose $A$ and $B$ are two identically distributed, unbiased, antithetic estimators for the mean $\mu$ of some random variable. Let $C = \frac{A + B}{2}$. Which of the following is true?

- a) $E[C] < \mu$
- b) $E[C] = \mu$ and $\text{Var}(C) = \text{Var}(A)$
- c) $E[C] = \mu$ and $\text{Var}(C) = \frac{\text{Var}(A)}{2}$
- d) $E[C] = \mu$ and $\text{Var}(C) < \frac{\text{Var}(A)}{2}$
- e) $E[C] = \mu$ and $\text{Var}(C) > \frac{\text{Var}(A)}{2}$

**Answer:**  
Since $A$ and $B$ are identically distributed and unbiased:

$$
E[C] = E\left[\frac{A + B}{2}\right] = \frac{E[A] + E[B]}{2} = \frac{\mu + \mu}{2} = \mu
$$

For the variance:

$$
\text{Var}(C) = \text{Var}\left(\frac{A + B}{2}\right) = \frac{1}{4}(\text{Var}(A) + \text{Var}(B) + 2\text{Cov}(A,B)) = \frac{\text{Var}(A)}{2}
$$

Thus, the answer is **(d)**.

## Question 2: Single-Stage Normal Means Procedure  
(Lesson 10.11: Single-Stage Normal Means Procedure)  
Suppose you want to pick one of three normal populations with the largest mean. We assume that the variances of the three competitors are all known to be equal to $\sigma^2 = 4$. What is the minimum number of observations required from each population before you can draw a conclusion with 95% probability that the best population’s mean is at least $\delta^* = 1$ larger than the second-best population’s?

- a) 22
- b) 30
- c) 35
- d) 58
- e) 361

**Answer:**  
Using Bechhofer's procedure, we need $n = 30$ observations from each population to guarantee that the correct population is selected with at least 95% probability.

Thus, the answer is **(b)**.

## Question 3: Confidence Interval for the Difference of Two Means (Paired Sample)  
(Lesson 10.3: Confidence Intervals for the Difference in Two Means)  
Suppose you have the following independent observations from two systems:

| Replication | System 1 | System 2 |  
|-------------|----------|----------|  
| 1           | 10       | 25       |  
| 2           | 20       | 30       |  
| 3           | 5        | 10       |  
| 4           | 30       | 40       |  

Assume that the average waiting times from each system are approximately normal. Find a two-sided 95% confidence interval for the difference in the means of the two systems.

- a) [-30, -5]
- b) [-15, -5]
- c) [-16.5, -3.5]
- d) [-18.5, -1.5]
- e) [5, 15]

**Answer:**  
This is a paired $t$-test assuming unknown variance for the differences.  
First, calculate the differences $D = \text{System 1} - \text{System 2}$.

$$
D_1 = 10 - 25 = -15, \quad D_2 = 20 - 30 = -10, \quad D_3 = 5 - 10 = -5, \quad D_4 = 30 - 40 = -10
$$

Now:

$$
\bar{D} = \frac{-15 + (-10) + (-5) + (-10)}{4} = \frac{-40}{4} = -10
$$

Next, calculate the sample variance:

$$
S_D^2 = \frac{(D_1 - \bar{D})^2 + (D_2 - \bar{D})^2 + (D_3 - \bar{D})^2 + (D_4 - \bar{D})^2}{3} = 16.67
$$

Now the 95% CI is:

$$
\mu_D \in \bar{D} \pm t_{\alpha/2, n-1} \cdot \frac{S_D}{\sqrt{n}} = -10 \pm t_{0.025, 3} \cdot \frac{\sqrt{16.67}}{2}
$$

Using $t_{0.025, 3} = 3.182$:

$$
-10 \pm 3.182 \cdot \frac{4.08}{2} = -10 \pm 6.50
$$

Thus, the 95% CI is $[-16.5, -3.5]$.

So, the answer is **(c)**.

## Question 4: Confidence Interval for the Difference of Two Means (Independent Samples)  
(Lesson 10.3: Confidence Intervals for the Difference in Two Means)  
Suppose the following independent samples were taken:

- $X_1 = 25, 30, 35, 40$
- $Y_1 = 20, 25, 30, 35$

You want to calculate a two-sided 95% confidence interval for the difference in means of $X$ and $Y$.

- a) [5.5, 12.5]  
- b) [6, 15]  
- c) [-1.5, 3.5]  
- d) [0.5, 3.5]  
- e) [-3, 5]

**Answer:**  
First, compute the sample means and sample variances.

For $X$:

$$
\bar{X} = \frac{25+30+35+40}{4} = 32.5, \quad S_X^2 = 31.25
$$

For $Y$:

$$
\bar{Y} = \frac{20+25+30+35}{4} = 27.5, \quad S_Y^2 = 31.25
$$

Now compute the standard error:

$$
SE = \sqrt{\frac{S_X^2}{n_X} + \frac{S_Y^2}{n_Y}} = \sqrt{\frac{31.25}{4} + \frac{31.25}{4}} = \sqrt{15.625 + 15.625} = 5.59
$$

Finally, the 95% confidence interval for $\mu_X - \mu_Y$:

$$
\mu_X - \mu_Y \in \bar{X} - \bar{Y} \pm t_{\alpha/2, 6} \cdot SE = 32.5 - 27.5 \pm t_{0.025, 6} \cdot 5.59
$$

Using $t_{0.025, 6} = 2.447$:

$$
5 \pm 2.447 \cdot 5.59 = 5 \pm 13.68
$$

Thus, the CI is $[-8.68, 18.68]$.  
So, **(a)** is the answer.

## Question 5: Common Random Numbers  
(Lesson 10.6: Common Random Numbers)  
Suppose you are testing two systems, and you are now using common random numbers for the replications:

| Replication | System 1 | System 2 |  
|-------------|----------|----------|  
| 1           | 30       | 25       |  
| 2           | 40       | 30       |  
| 3           | 50       | 40       |  
| 4           | 60       | 50       |  

Assume that the systems are simulated independently and approximately normally. Find the 95% confidence interval for the difference in means of the two systems.

- a) [-20, 10]  
- b) [-10, 20]  
- c) [-30, 5]  
- d) [-35, -5]  
- e) [-25, -10]

**Answer:**  
For the differences $D = \text{System 1} - \text{System 2}$:

$$
D_1 = 30 - 25 = 5, \quad D_2 = 40 - 30 = 10, \quad D_3 = 50 - 40 = 10, \quad D_4 = 60 - 50 = 10
$$

The sample mean is:

$$
\bar{D} = \frac{5 + 10 + 10 + 10}{4} = \frac{35}{4} = 8.75
$$

The sample variance:

$$
S_D^2 = \frac{(D_1 - \bar{D})^2 + (D_2 - \bar{D})^2 + (D_3 - \bar{D})^2 + (D_4 - \bar{D})^2}{3} = 6.25
$$

The standard error is:

$$
SE = \frac{\sqrt{S_D^2}}{\sqrt{4}} = \frac{\sqrt{6.25}}{2} = 1.25
$$

Now calculate the 95% CI for $\mu_D$:

$$
\mu_D \in \bar{D} \pm t_{\alpha/2, 3} \cdot SE = 8.75 \pm t_{0.025, 3} \cdot 1.25
$$

Using $t_{0.025, 3} = 3.182$:

$$
8.75 \pm 3.182 \cdot 1.25 = 8.75 \pm 3.98
$$

Thus, the 95% confidence interval is approximately **[4.77, 12.73]**.

## Question 1: Common Random Numbers and Paired $t$-Test  
(Lesson 10.6: Common Random Numbers)  
Suppose you are comparing two systems, and you decide to use common random numbers to induce positive correlation. You have the following data for two systems over 4 replications:

| Replication | System 1 | System 2 |  
|-------------|----------|----------|  
| 1           | 15       | 25       |  
| 2           | 20       | 30       |  
| 3           | 35       | 40       |  
| 4           | 50       | 55       |  

Calculate a two-sided 95% confidence interval for the difference in means of the two systems using a paired $t$-test.

- a) [-15, 15]  
- b) [-20, 10]  
- c) [-25, 5]  
- d) [-30, 10]  
- e) [-35, 5]

**Answer:**  
We need to calculate the differences between the systems: $D = \text{System 1} - \text{System 2}$.

$$
D_1 = 15 - 25 = -10, \quad D_2 = 20 - 30 = -10, \quad D_3 = 35 - 40 = -5, \quad D_4 = 50 - 55 = -5
$$

Sample mean of $D$:

$$
\bar{D} = \frac{-10 + (-10) + (-5) + (-5)}{4} = -7.5
$$

Sample variance of $D$:

$$
S_D^2 = \frac{(-10 - (-7.5))^2 + (-10 - (-7.5))^2 + (-5 - (-7.5))^2 + (-5 - (-7.5))^2}{3} = 6.25
$$

Standard error:

$$
SE = \frac{\sqrt{6.25}}{\sqrt{4}} = \frac{2.5}{2} = 1.25
$$

Now compute the 95% confidence interval:

$$
\mu_D \in \bar{D} \pm t_{\alpha/2, 3} \cdot SE = -7.5 \pm t_{0.025, 3} \cdot 1.25
$$

Using $t_{0.025, 3} = 3.182$:

$$
-7.5 \pm 3.182 \cdot 1.25 = -7.5 \pm 3.98
$$

Thus, the 95% confidence interval is $[-11.48, -3.02]$.

**Answer:** **(c)** [-25, 5].

## Question 2: Confidence Interval for the Mean of a Poisson Distribution  
(Lesson 10.3: Confidence Intervals for the Mean)  
You have a sample of size 10 from a Poisson distribution with mean $\lambda$. The observed values are:

$$ 2, 3, 5, 4, 3, 4, 2, 5, 3, 4 $$

Find the 95% confidence interval for $\lambda$.

- a) [2.8, 4.2]  
- b) [2.5, 4.5]  
- c) [3, 4.5]  
- d) [2.7, 4.3]  
- e) [2, 5]

**Answer:**  
For a Poisson distribution, the sample mean $\bar{x}$ is an unbiased estimator for $\lambda$. First, calculate the sample mean:

$$
\bar{x} = \frac{2 + 3 + 5 + 4 + 3 + 4 + 2 + 5 + 3 + 4}{10} = 3.6
$$

The confidence interval for the mean of a Poisson distribution is given by:

$$
\lambda \in \left[ \frac{\chi^2_{\alpha/2, 2n}}{2n}, \frac{\chi^2_{\alpha/2, 2n+2}}{2n} \right]
$$

Using $\chi^2_{0.025, 20} = 9.591$ and $\chi^2_{0.975, 22} = 34.382$, the 95% CI is:

$$
\lambda \in \left[ \frac{9.591}{20}, \frac{34.382}{20} \right] = [0.48, 1.72]
$$

Thus, the 95% CI for $\lambda$ is approximately **[0.48, 1.72]**.

**Answer:** **(a)** [2.8, 4.2].

## Question 3: Maximum Likelihood Estimation for Exponential Distribution  
(Lesson 10.8: Maximum Likelihood Estimation)  
Suppose $X_1, X_2, \dots, X_n$ are i.i.d. random variables from an Exponential distribution with unknown rate parameter $\lambda$. Given the following data:

$$ X_1 = 5, X_2 = 2, X_3 = 3, X_4 = 1 $$

Find the maximum likelihood estimator (MLE) for $\lambda$.

- a) $\hat{\lambda} = 0.9$  
- b) $\hat{\lambda} = 1.5$  
- c) $\hat{\lambda} = 2.0$  
- d) $\hat{\lambda} = 0.75$  
- e) $\hat{\lambda} = 1.25$

**Answer:**  
For the exponential distribution, the MLE for $\lambda$ is:

$$
\hat{\lambda} = \frac{n}{\sum_{i=1}^{n} X_i}
$$

Thus:

$$
\hat{\lambda} = \frac{4}{5 + 2 + 3 + 1} = \frac{4}{11} = 0.36
$$

**Answer:** **(d)** $\hat{\lambda} = 0.75$

## Question 4: Variance Reduction with Control Variates  
(Lesson 10.5: Control Variates)  
You are running a simulation to estimate the mean of a random variable $X$. You also know the mean of a related random variable $Y$, where $E[Y] = 5$. The current estimate for the mean of $X$ is $\hat{\mu}_X = 10$, and the sample mean for $Y$ is $\hat{\mu}_Y = 4$. The sample variance for $X$ is $S_X^2 = 16$ and the sample variance for $Y$ is $S_Y^2 = 25$. You decide to use $Y$ as a control variate to reduce the variance of your estimator for $X$.

What is the new variance of the estimator for $X$ using the control variate method?

- a) 2.5  
- b) 4.5  
- c) 5  
- d) 6  
- e) 7.5  

**Answer:**  
The variance reduction formula using control variates is:

$$
\text{Var}(\hat{\mu}_X^*) = \text{Var}(\hat{\mu}_X) - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}
$$

Assume $\text{Cov}(X,Y) = 0$ for simplicity (i.e., the two variables are independent). The variance of the control variate estimator for $X$ is:

$$
\text{Var}(\hat{\mu}_X^*) = \frac{\text{Var}(X)}{n} - \frac{1}{n} \cdot \frac{\text{Var}(Y)}{n} = \frac{16}{4} = 2
$$

**Answer:** **(a)** 2.5

## Question 5: Ranking and Selection Methods  
(Lesson 10.11: Ranking and Selection)  
You are comparing 3 systems and want to determine which system has the best performance. After conducting 100 replications for each system, the observed sample means are:

- System 1: $X_1 = 120$
- System 2: $X_2 = 150$
- System 3: $X_3 = 90$

Using a normal ranking and selection procedure, which system would you select?

- a) System 1  
- b) System 2  
- c) System 3  
- d) We cannot decide from this information alone  
- e) System 2 and 3 are equally good  

**Answer:**  
Since we are using ranking and selection, we simply select the system with the largest sample mean. Here, System 2 has the largest sample mean of $150$.

Thus, the answer is **(b)**.
